```python
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA,KernelPCA
from sklearn.manifold import LocallyLinearEmbedding,MDS,TSNE,Isomap
from sklearn.random_projection import GaussianRandomProjection,SparseRandomProjection
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as pl
import numpy as np
```

```python
n_samples = 1500
noise = 0.05
X, color = make_swiss_roll(n_samples=n_samples, noise=noise)
```

```python
print(X)
```

```python
# Plot the swiss roll dataset
def visualize(X):
    fig = pl.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=pl.cm.Spectral)
    ax.set_title("Swiss Roll Dataset")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    pl.show()
visualize(X)
```

```python
def visualize_compressed(X):
    pl.scatter(X[:, 0], X[:, 1], c=color, cmap=pl.cm.Spectral)
    pl.title("Swiss Roll Dataset Projected to 2D")
    pl.colorbar(label='Color (Position along the roll)')
    pl.show()
```

# Using PCA

```python
pca=PCA(n_components=2)
x1=pca.fit_transform(X)
print('Using PCA')
visualize_compressed(x1)
```

# Using Random Projection

```python
grp=GaussianRandomProjection(n_components=2,random_state=42)
x2=grp.fit_transform(X)
print('Gaussian Random Projection')
visualize_compressed(x2)
srp=SparseRandomProjection(n_components=2,random_state=42)
x3=srp.fit_transform(X)
print('Sparse Random Projection')
visualize_compressed(x2)
```

# Using Locally Linear Embedding

```python
lle=LocallyLinearEmbedding(n_neighbors=20,n_components=2,random_state=42)
x4=lle.fit_transform(X)
print('Locally Linear Embedding')
visualize_compressed(x4)
```

# Using MDS

```python
mds = MDS(n_components=2,normalized_stress=False,random_state=42)
x5=mds.fit_transform(X)
print('Multi-Dimensional Scaling')
visualize_compressed(x5)
```

# Using Isomap

```python
im = Isomap(n_components=2)
x6=im.fit_transform(X)
print('Isomap')
visualize_compressed(x6)
```

# Using TSNE

```python
tsne = TSNE(n_components=2,random_state=42)
x7=tsne.fit_transform(X)
print('T-Distributed Stochastic Neighbor Embedding')
visualize_compressed(x7)
```
